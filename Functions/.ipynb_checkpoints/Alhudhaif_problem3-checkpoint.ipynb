```python
import pandas as pd
import numpy as np
import scipy.optimize as opt
import time
```

```python
# Read in the dataset
set3 = pd.read_stata('PS3_data.dta')
set3.head(n=5)
```

```python
set3 = set3[(set3.hlabinc.isnull() != True)&(set3.hannhrs.isnull() != True) & (set3.hannhrs > 0) &
           (set3.hyrsed.isnull() != True) & (set3.age.isnull() != True) & 
            (set3.hrace.isnull() != True)]
```

```python
set3['wageperhour'] = set3['hlabinc'] / set3['hannhrs']
set3['logwage']= np.log(set3['wageperhour'])
```

```python
set3= set3[(set3.hsex==1)&(set3.age >= 25)&(set3.age <= 60)&(set3.wageperhour > 7)]
```

```python
set3.describe()
```

```python
set3['Black'] = set3.loc[:,'hrace'].apply(lambda x : 1 if x == 2 else 0)
set3['Hispanic'] = set3.loc[:,'hrace'].apply(lambda x : 1 if x == 5 else 0)
set3['OtherRace'] = set3.loc[:,'hrace'].apply(lambda x : 1 if x == 3 or x == 4 or x == 6 or x == 7 else 0)
```

```python
set3['cons']=1
```

# Adding a year as a parameter:

```python
# Assuming normal distribution of the error term, the log-likelihood function 
#is defined via the following:
    
def loglike(params, year, dataset):
    """This function takes the tuple of parameters, the given year, and the name of the dataset, and returns the 
    log likelihood value:
    parameters: a tuple of the coeffecients
    year: limit the observations to be for a given year, in order to calculate the log likelihood value for that year
    dataset: the name of the dataset being used """
    
    z=np.array([list(params)])
    beta = np.transpose(z)
    
    #n returns the sum of obs that have a value equal to the parameter year
    n = np.sum(dataset.year == year)
    
    #The following [dataset['cons'][dataset.year==year] >>>> This extract a portion of the data set that has a given year
    
    
    x = np.array([dataset['cons'][dataset.year == year], dataset['hyrsed'][dataset.year == year],
                  dataset['age'][dataset.year == year], dataset['Black'][dataset.year == year], 
                  dataset['OtherRace'][dataset.year == year]]).T
    
    y = np.array([dataset['logwage'][dataset.year == year]]).T
    
    epsilon = y - np.dot(x, beta)
    
    essilon_s = np.dot(epsilon, epsilon.T)
    
    # the varaiance of the error term is measured by the function "error"
    sigma = error(dataset, year, params)
    
    
    
    
   
    
    log_like = (((n/2) * np.log(2 * (np.pi))) + ((n)/2)* np.log((sigma**2)) +
                  (( essilon_s / ((sigma ** 2))/2))).item(0,0)
    
    return log_like
```

```python
def error(dataset, year, params):
    """This function returns the variance of the error term, it requires the following inputs:
    dataset: the dataset being used
    year: constrain the obs to be for a given year
    params: the error term is measured after specifying a tuple of coeffecients
    """
    z=np.array([list(params)])
    beta = np.transpose(z)
    
    #The following [dataset['cons'][dataset.year==year] >>>> This extract a portion of the data set that has a given year
    
    n = np.sum(dataset.year == year)
    x = np.array([dataset['cons'][dataset.year == year], dataset['hyrsed'][dataset.year == year],
                  dataset['age'][dataset.year == year], dataset['Black'][dataset.year == year], 
                  dataset['OtherRace'][dataset.year == year]]).T
    
    y = np.array([dataset['logwage'][dataset.year == year]]).T
    
    var = np.dot((y - np.dot(x, beta)).T, (y - np.dot(x, beta)) ) / n
                 

    return var.item(0,0)
```

# Finding MLE coeffecients for year 1971:

```python
params0 = (0.1, -0.1, 0.2, -0.1, 0.6)
bnds = ((None, None), (None, None), (None, None),
        (None, None), (None, None))
year=1971
dataset=set3
# Call minimzer to minimize the log-likelihood function
mle_results = opt.minimize(loglike, params0,args=(year, dataset),
                           method='SLSQP',bounds=bnds, tol=1e-15,)

mle_results
```

# Finding MLE coeffecients for year 1971, 1980, and 1990 :

```python
years=[1971, 1980, 1990]
for i in years:
    year= i
    params0 = (0.1, -0.1, 0.2, -0.1, 0.6)
    bnds = ((None, None), (None, None), (None, None),
        (None, None), (None, None))
    dataset=set3
    
    # Call minimzer to minimize the log-likelihood function
    mle_results = opt.minimize(loglike, params0,args=(year, dataset),
                           method='SLSQP',bounds=bnds, tol=1e-15,)

    print('For year '+str(i)+':'+str(mle_results))
```
